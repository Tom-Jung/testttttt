```python
import numpy as np
import pandas as pd
```

```python
Ad = pd.read_csv('/content/Rg_io.csv') # Colab폴더에서 Pandas의 DataFrame으로 불러오기
Ad.head()
```

```python
Add = Ad.to_numpy()
```

```python
Lf = np.linalg.inv(np.eye(561)-Add)
```

```python
df_Lf = pd.DataFrame(Lf)
df_Lf
```

```python
df_Lf.to_csv('지역생산유발계수.csv', encoding="utf-8-sig")
```

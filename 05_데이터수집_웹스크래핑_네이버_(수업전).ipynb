6) 전체코드

```python
import requests
from bs4 import BeautifulSoup
url="https://n.news.naver.com/mnews/article/001/0014581380?sid=101"

#BeautifulSoup로 html스크래핑
resp=requests.get(url)
soup=BeautifulSoup(resp.text)

#뉴스제목 스크래핑
selector='#title_area > span'
news_title=soup.select(selector)
print('뉴스제목:', news_title[0].text)

#기자이름 스크래핑
selector='#ct > div.media_end_head.go_trans > div.media_end_head_info.nv_notrans > div.media_end_head_journalist > a > em'
reporter_name=soup.select(selector)
print('기자이름:',reporter_name[0].text)

#기사내용 스크래핑
selector='#dic_area'
news_contents=soup.select(selector)
print('기사내용:',end='')
for line in news_contents:
    text=line.get_text(separator='\n')
    print(text)
```
